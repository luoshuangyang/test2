```python
from __future__ import unicode_literals

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager
from matplotlib.font_manager import FontProperties 

font = FontProperties(fname=r"/root/anaconda2/envs/python3/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/msyh.ttf")

import numpy as np
import scipy.stats as s

from scipy.optimize import least_squares, curve_fit
```

```python
data1 = pd.read_csv("3505945.csv", encoding = "GB2312")
data1.columns
```

```python
data1 = pd.read_csv("3505945.csv", encoding = "GB2312")
data1["部件本次装机使用小时"]=data1["部件本次装机使用小时"].str.split(':').str[0].astype(int)
data_h=data1[data1["拆换原因"]=="FAIL  "]
data_h=data_h[data_h["部件装上使用循环数"]==0]
data_h["部件本次装机使用小时"].hist(bins=9, range=(0,20000))
plt.ylabel("Frequency")
plt.xlabel("Time to failure")
plt.title('Histogram of Time to failure')
```

```python
plt.scatter(data_h.index,data_h['部件本次装机使用小时'])
plt.ylabel("Time to failure")
plt.xlabel("Observation")
plt.title('Run Chart of Time to failure')
```
