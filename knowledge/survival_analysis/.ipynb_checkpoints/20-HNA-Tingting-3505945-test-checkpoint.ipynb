```python
x0 = np.array([0.01,0.01])

def model(x, u):
    return x[1] / (x[0]**x[1]) *(u**(x[1]-1)) * np.exp(-((u/x[0])**x[1]))

def fun(x, u, y):
    return model(x, u) - y
res = least_squares(fun,x0, bounds=(0, 100), args=(bins, values), verbose=1)
print(res.x)
```

```python
u_test = np.linspace(0, 20000)
y_test = model(res.x, u_test)
print(u_test.shape)
print(y_test)
plt.plot(bins,values,'o', markersize=4, label='data')
plt.plot(u_test, y_test, label='fitted model')
```
