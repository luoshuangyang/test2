生存分析是统计学的一个分支，用于分析一个或多个事件发生前的预期持续时间，如生物有机体的死亡和机械系统的故障。本课题在工程上称为可靠性理论或可靠性分析，在经济学上称为持续时间分析或持续时间建模，在社会学上称为事件历史分析。

#### 生存分析试图回答这样的问题:在一定时间内存活下来的人口比例是多少?

那些幸存下来的，他们将以什么速度死亡或失败?是否可以考虑死亡或失败的多种原因?特定的环境或特征如何增加或减少生存的可能性?为了回答这些问题，有必要对“一生”下一个定义。在生物生存的情况下，死亡是明确的，但就机械可靠性而言，失败可能没有明确的定义，因为很可能有机械系统中的失败是部分的，是一个程度的问题，或者不是局部的。即使在生物学问题中，某些事件(例如，心脏病发作或其他器官衰竭)也可能具有相同的模糊性。下面概述的理论假设在特定的时间发生定义明确的事件;其他情况可以用模型更好地处理，模型明确地解释了不明确的事件。更一般地说，生存分析包括时间到事件数据的建模;在这种情况下，死亡或失败在生存分析文献中被认为是一个“事件”，传统上每个研究对象只发生一个事件，之后机体或机制死亡或破坏。重复事件或重复事件模型放松了这种假设。重复事件的研究与系统可靠性相关，在社会科学和医学研究的许多领域也有关系。
